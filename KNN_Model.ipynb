# Creating a K-Nearest Neighbours (KNN) model from 0

## Setting environment

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from collections import Counter
from sklearn.model_selection import train_test_split
```

## Creating the basses

```python
df = pd.read_csv('Iris.csv')
X = df.drop(columns=['Id', 'Species'])
y = df['Species']

## Here it's important to note that is best practice to standarize the values, because
## large-measure values could overweight other more defining values

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_standarized = scaler.fit_transform(X)
```

```python
def euclideian_distance(a,b):
    return np.sqrt(np.sum((b-a)**2))
## This is a basic concept of algebra, which comes from the Pithagorean Theorem
## Basically, you substract the values of a dimension to both points, to later sum their squares
## At last, you place the value to a square root
```

```python
class KNN():
   def __init__(self, k):
      self.k = k

   def fit(self, X, y):
      self.X = np.array(X)
      self.y = np.array(y)

   def predict(self, new_point):
      predictions = (self.predict_class(points) for points in new_point)
      return list(predictions)
   
   def predict_class(self, point):
      distance = np.array([euclideian_distance(point, value) for value in self.X])
      ## As the list of values follows the same order as the list of X values, we can extract their indices
      nearest_index = np.argsort(distance)[:self.k]
      ## Here we are not only extracting the indices of tghe nearest values amd sorting them by smallest to largest distance, but also retucing the list to the first k-values
      nearest_labels = [self.y[i] for i in nearest_index]
      most_common = Counter(nearest_labels).most_common(1)[0][0]
      return most_common
```

## Now that we have already constructed the model, and as this model can't be trained, lets directly test it

```python
X_train, X_test, y_train, y_test = train_test_split(X_standarized, y, test_size=0.2, random_state=990)
```

```python
print("Train:", Counter(y_train))
print("Test:", Counter(y_test))
```

### As we can see, I had an incredibly rare encounter with my "random_state" selection, as it took evenly from the three categories without me using the stratification (a method of the train_test_split() function which always represents all posible variables in order to eliminate cases when one category doesn't appear in the "training" data).

### For studying reasons, I am gonna change the rqandom_state, eventhough is important to clarify that using stratification is almoust always recommended in classification models.

```python
X_train, X_test, y_train, y_test = train_test_split(X_standarized, y, test_size=0.2, random_state=42, stratify=y)
```

```python
KNN_model = KNN(30)
KNN_model.fit(X_train, y_train)
predictions = KNN_model.predict(X_test)
```

```python
accuracy = np.mean(np.array(predictions) == y_test)*100
print(f'The accruracy is {accuracy:.2f}%')
```

```python
KNN_model_2 = KNN(1)
KNN_model_2.fit(X_train, y_train)
predictions_2 = KNN_model_2.predict(X_test)
accuracy_2 = np.mean(np.array(predictions_2) == y_test)*100
print(f'The accruracy is {accuracy_2:.2f}%')
```

### In this case, the choosing of K isn't as crucial as it usually is, mainly because this particular data set is really stable and constant, whith categories clearly separate.
